```python
import numpy as np

# Sample data
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Parameters
w0 = 0.0
w1 = 0.0
alpha = 0.01
m = len(y)
iterations = 1000

# Hypothesis function h = w0 + w1 * x
def predict(x, w0, w1):
    print(f"x: {x}, w1*x= {w1*x}, w0+w1*x= {w0+w1*x}")
    p = w0 + w1 * x
    print(f"p: {p}")
    return p

# Cost function
def cost_function(x, y, w0, w1):
    predictions = predict(x, w0, w1)
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    print(f"cost: {cost}")
    return cost

# Final parameters
print(f"w0: {w0}, w1: {w1}")
print(f"Final cost: {cost_function(x, y, w0, w1)}")
```
